```python
df = spark.read.format("delta").load("abfss://WSP_MartinicaRealState@onelake.dfs.fabric.microsoft.com/LK_brz_MartinicaRealState.Lakehouse/Tables/brokers")
df.createOrReplaceTempView("tmp_bro")

df = spark.read.format("delta").load("abfss://WSP_MartinicaRealState@onelake.dfs.fabric.microsoft.com/LK_brz_MartinicaRealState.Lakehouse/Tables/campaigns")
df.createOrReplaceTempView("tmp_cam")

df = spark.read.format("delta").load("abfss://WSP_MartinicaRealState@onelake.dfs.fabric.microsoft.com/LK_brz_MartinicaRealState.Lakehouse/Tables/clients")
df.createOrReplaceTempView("tmp_cli")

df = spark.read.format("delta").load("abfss://WSP_MartinicaRealState@onelake.dfs.fabric.microsoft.com/LK_brz_MartinicaRealState.Lakehouse/Tables/projects")
df.createOrReplaceTempView("tmp_prj")

df = spark.read.format("delta").load("abfss://WSP_MartinicaRealState@onelake.dfs.fabric.microsoft.com/LK_brz_MartinicaRealState.Lakehouse/Tables/properties")
df.createOrReplaceTempView("tmp_pro")
```

```python
%%sql
MERGE INTO brokers AS tgt
USING tmp_bro AS src
ON tgt.BrokerID = src.BrokerID
WHEN MATCHED THEN
  UPDATE SET
    tgt.BrokerName = src.BrokerName,
    tgt.Region = src.Region,
    tgt.Email = src.Email
WHEN NOT MATCHED THEN
  INSERT (BrokerID, BrokerName, Region, Email)
  VALUES (src.BrokerID, src.BrokerName, src.Region, src.Email);
```

```python
%%sql
MERGE INTO campaigns AS TG
USING tmp_cam AS OS
ON TG.CampaignID = OS.CampaignID
WHEN MATCHED THEN
  UPDATE SET
    TG.Channel     = OS.Channel,
    TG.CampaignName= OS.CampaignName,
    TG.StartDate   = OS.StartDate,
    TG.EndDate     = OS.EndDate,
    TG.BudgetUSD   = OS.BudgetUSD
WHEN NOT MATCHED THEN
  INSERT (CampaignID, Channel, CampaignName, StartDate, EndDate, BudgetUSD)
  VALUES (OS.CampaignID, OS.Channel, OS.CampaignName, OS.StartDate, OS.EndDate, OS.BudgetUSD);
```

```python
%%sql
MERGE INTO clients AS T
USING tmp_cli AS S
ON T.ClientID = S.ClientID
WHEN MATCHED THEN
  UPDATE SET
    T.FirstName = S.FirstName,
    T.LastName  = S.LastName,
    T.Email     = S.Email,
    T.Region    = S.Region
WHEN NOT MATCHED THEN
  INSERT (ClientID, FirstName, LastName, Email, Region)
  VALUES (S.ClientID, S.FirstName, S.LastName, S.Email, S.Region);
```

```python
%%sql
MERGE INTO projects AS TG
USING tmp_prj AS OS
ON TG.ProjectID = OS.ProjectID

WHEN MATCHED THEN
  UPDATE SET
    TG.ProjectName = OS.ProjectName,
    TG.City        = OS.City,
    TG.Region      = OS.Region,
    TG.LaunchYear  = OS.LaunchYear,
    TG.Status      = OS.Status

WHEN NOT MATCHED THEN
  INSERT (ProjectID, ProjectName, City, Region, LaunchYear, Status)
  VALUES (OS.ProjectID, OS.ProjectName, OS.City, OS.Region, OS.LaunchYear, OS.Status);
```

```python
%%sql
MERGE INTO properties AS T
USING tmp_pro AS S
ON T.PropertyID = S.PropertyID
WHEN MATCHED THEN UPDATE SET
    T.ProjectID = S.ProjectID,
    T.PropertyType = S.PropertyType,
    T.Size_m2 = S.Size_m2,
    T.Bedrooms = S.Bedrooms,
    T.Bathrooms = S.Bathrooms,
    T.ListPriceUSD = S.ListPriceUSD,
    T.AvailabilityStatus = S.AvailabilityStatus
WHEN NOT MATCHED THEN INSERT (
    PropertyID,
    ProjectID,
    PropertyType,
    Size_m2,
    Bedrooms,
    Bathrooms,
    ListPriceUSD,
    AvailabilityStatus
) VALUES (
    S.PropertyID,
    S.ProjectID,
    S.PropertyType,
    S.Size_m2,
    S.Bedrooms,
    S.Bathrooms,
    S.ListPriceUSD,
    S.AvailabilityStatus
);
```
